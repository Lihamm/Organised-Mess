#### Looking at the price of each coin

```python
df8 = df[['name', 'quote.USD.price', 'timestamp']]
df8
```

Query the df for a particular coin e.g. Bitcoin

```python
df9 = df8.query("name == 'Bitcoin'")
df9 = df9.set_index('name')
df9 = df9.reset_index()
df9
```

```python
sns.lineplot(x='timestamp', y='quote.USD.price', data=df9)
plt.xticks(rotation=45)
```

Reducing the time stamp to datetime

```python
df10 = df9['timestamp'].str.split(' ', expand=True)
df11 = df9.join(df10, how='right')
df11
```

Now because I only want time I will remove columns 'timestamp' and '0' and rename column '1' as 'time

```python
df12 = df11.drop(['timestamp', 0], axis=1)
df12
```

Seperate the time by delimeter '.'

```python
df13 = df11[1].str.split('.', expand=True)
df14 = df13.drop([1], axis=1)
df14
```

```python
df15 = df12.join(df14, how='right')
df15
```

```python
df16 = df15.drop([1], axis=1)
df16.rename(columns={0:'time'}, inplace=True)
df16
```

```python
sns.lineplot(x='time', y='quote.USD.price', data=df16)
plt.xticks(rotation=45)
sns.set_theme(style="darkgrid")
```
